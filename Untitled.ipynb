```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bus_data['type'] = 'bus'
```

```python
subway_data['type'] = 'subway'
```

```python
merged_data = pd.concat([subway_data, bus_data], ignore_index=True)
```
